* **Analyse an Image using Histogram**

```python
car_img = cv.imread("W140.jpg")


"""
[0], [1], [2]: Channel indices (0=Blue, 1=Green, 2=Red in BGR format)
None: No mask applied (process entire image)
[256]: Number of bins (256 for 8-bit images, covering 0-255)
[0, 256]: Pixel value range

"""

hist_img = cv.calcHist([car_img], [0], None, [255], [0, 255])

plt.plot(hist_img, color='b')
plt.title("GrayScale Histogram")
plt.xlabel("Pixel value")
plt.ylabel("Frequency")
plt.show()
```

```python
car_img = cv.imread("W140.jpg")

gray_hist_img = cv.calcHist([car_img], [0], None, [256], [0, 256])
hist_img_1 = cv.calcHist([car_img], [1], None, [256], [0, 256])
hist_img_2 = cv.calcHist([car_img], [2], None, [256], [0, 256])

plt.plot(gray_hist_img, color='b')
plt.plot(hist_img_1, color='r')
plt.plot(hist_img_2, color='k')
plt.title("DifferentScale of Histogram")
plt.xlabel("Pixel value")
plt.ylabel("Frequency")
plt.show()
```

* **Mouse Event Bindings in OpenCV**

```python
def mouse_callback(event, x, y, flags, param):
    if event == cv.EVENT_LBUTTONDOWN:
        print(f"Left click at : {x}, {y}")
    elif event == cv.EVENT_RBUTTONDOWN:
        print(f"Right click at : {x}, {y}")

img = cv.imread("Motor.jpg")
cv.imshow("Mouse Event", img)
cv.setMouseCallback("Mouse Event", mouse_callback)
cv.waitKey(0)
cv.destroyAllWindows()
```

* **Image Contours**

```python
image_w140 = cv.imread("W140.jpg")
gray_img = cv.cvtColor(image_w140, cv.COLOR_BGR2GRAY)
_, thresh = cv.threshold(gray_img, 100, 255, cv.THRESH_BINARY)
contours, _ = cv.findContours(thresh, cv.RETR_TREE, cv.CHAIN_APPROX_SIMPLE)

cv.drawContours(image_w140, contours, -1, (0, 0, 150), 1)
cv.imshow("Contours", image_w140)
cv.waitKey(0)
cv.destroyAllWindows()
```

* **Object Detection using Contours by Webcam**

```python
cap = cv.VideoCapture(0)

while cap.isOpened:
    ret, frame = cap.read()
    gray = cv.cvtColor(frame, cv.COLOR_BGR2GRAY)
    _, thresh = cv.threshold(gray, 150, 255, cv.THRESH_BINARY)
    contours, _ = cv.findContours(thresh, cv.RETR_TREE, cv.CHAIN_APPROX_SIMPLE)

    cv.drawContours(frame, contours, -2, (0, 0, 150), 1)
    cv.imshow("Contours Live", frame)

    if cv.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv.destroyAllWindows()
```

* **Image Background Removal with thresh**

```python
img = cv.imread("W140.jpg")
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
_, thresh = cv.threshold(gray, 110, 255, cv.THRESH_BINARY_INV)


result = cv.bitwise_and(img, img, mask=thresh)
cv.imshow('Removed Background', result)
cv.waitKey(0)
cv.destroyAllWindows()
```

* **Hough Circle Detection**

```python
image = cv.imread("Motor.jpg", cv.IMREAD_COLOR)
gray = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
gray = cv.medianBlur(gray, 3)

circles = cv.HoughCircles(
    gray, cv.HOUGH_GRADIENT, 1, 50, param1=10, param2=20, minRadius=0, maxRadius=0
)

if circles is not None:
    circles = np.uint16(np.around(circles))
    for i in circles[0, :]:
        cv.circle(image, (i[0], i[1]), i[2], (0, 255, 0), 2)
        cv.circle(image, (i[0], i[1]), 2, (0, 0, 255), 2)


cv.imshow("Hough Circles", image)
cv.waitKey(0)
cv.destroyAllWindows()
```
