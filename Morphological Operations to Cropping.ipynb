* **Morphological Operations Part1 (Erosion & Dilation)**

```python
img = cv.imread("F22_Raptor.jpg", 0)            # Color ====> Gray

_, mask = cv.threshold(img, 100, 50, cv.THRESH_BINARY)

kernel = np.ones((3, 3), np.uint8)

erosion = cv.erode(mask, kernel, iterations=1)
dilation = cv.dilate(mask, kernel, iterations=1)

cv.imshow("Original image", img)
cv.imshow("Original (Mask)", mask)
cv.imshow("Erosion", erosion)
cv.imshow("Dilation", dilation)

cv.waitKey(0)
cv.destroyAllWindows()
```

* **Morphological Operations Part2 (Opening & Closing)**

```python
opening = cv.morphologyEx(mask, cv.MORPH_OPEN, kernel)
closing = cv.morphologyEx(mask, cv.MORPH_CLOSE, kernel)

cv.imshow("Opening", opening)
cv.imshow("Closing", closing)
cv.waitKey(0)
cv.destroyAllWindows()
```

* **Image Pyramid**

```python
image = cv.imread("F22_Raptor.jpg")

down = cv.pyrDown(image)
up = cv.pyrUp(image)

cv.imshow("Original", image)
cv.imshow("Downsampled", down)
cv.imshow("Upsampled", up)
cv.waitKey(0)
cv.destroyAllWindows()
```

* **Image Translation**

```python
car_img = cv.imread("W140.jpg")

rows, cols = car_img.shape[:2]

M = np.float32([[1, 0, 30], [0, 1, 150]])       # shift x by 150 and shift y by 200

translated = cv.warpAffine(car_img, M, (cols, rows))

cv.imshow("Original image", car_img)
cv.imshow("Translated image", translated)
cv.waitKey(0)
cv.destroyAllWindows()
```

* **Geometric Transformation**

```python
center = (cols // 2, rows // 2)
matrics = cv.getRotationMatrix2D(center, angle=45, scale=1)
rotated = cv.warpAffine(car_img, matrics, (cols, rows))

cv.imshow("Rotated", rotated)
cv.waitKey(0)
cv.destroyAllWindows()
```

* **Background Subtraction**

```python
cap = cv.VideoCapture("FSI.mp4")
fgbg = cv.createBackgroundSubtractorMOG2()

while cap.isOpened:
    ret, frame = cap.read()
    if not ret:
        break
    fgmask = fgbg.apply(frame)
    cv.imshow("Original Video", frame)
    cv.imshow("Background_Sub Mask Video", fgmask)
    if cv.waitKey(20) & 0xFF == ord('q'):
        break

cap.release()
cv.destroyAllWindows()
```

* **Extract Images from Video**

```python
cap = cv.VideoCapture("FSI.mp4")

count = 0

while cap.isOpened:
    ret, frame = cap.read()

    if not ret:
        break
    cv.imwrite(f"frame{count}.jpg", frame)
    count += 1

cap.release()
print(f"Extracted {count} frames")
```

* **Convert Colors**

```python
Motor_img = cv.imread("Motor.jpg")

gray_img = cv.cvtColor(Motor_img, cv.COLOR_BGR2GRAY)
hsv_img = cv.cvtColor(Motor_img, cv.COLOR_BGR2HSV)

cv.imshow("Original Motor image", Motor_img)
cv.imshow("Gray image", gray_img)
cv.imshow("HSV image", hsv_img)

cv.waitKey(0)
cv.destroyAllWindows()
```

* **Crop Image**

```python
main_img = cv.imread("W140.jpg")
cropped = main_img[150:400, 150:550]

cv.imshow("main image", main_img)
cv.imshow("cropped image", cropped)
cv.waitKey(0)
cv.destroyAllWindows()
```

* **Create Blank images (Black & White)**
