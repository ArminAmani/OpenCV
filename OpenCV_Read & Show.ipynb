* **Show Multiple Images and Slidshow**

```python
images = ['W140.jpg', 'Motor.jpg', 'F22_Raptor.jpg']

for img_path in images:

    # Check if file exists
    if not os.path.exists(img_path):        # if the file does NOT exist
        print(f"Error: file {img_path} not found!")
        continue
    
    img = cv.imread(img_path)

    # Check Loaded Successfully of image
    if img is None:
        print(f"Error: Could not load image '{img_path}'")
        continue
    
    cv.imshow("Slidshow", img)
    cv.waitKey(1000)
cv.destroyAllWindows()
```

* **Read and show the Images with Color**

```python
img_color = cv.imread("Motor.jpg", cv.IMREAD_COLOR)
img_gray = cv.imread("Motor.jpg", cv.IMREAD_GRAYSCALE)

cv.imshow("Color Image", img_color)
cv.imshow("GrayScale Image", img_gray)

cv.waitKey(0)
cv.destroyAllWindows()
```

* **Add text over an Image**

```python
image = cv.imread("W140.jpg")
font = cv.FONT_HERSHEY_COMPLEX

cv.putText(image, "W140-Benz(My ideal car)", (20, 30), font, 0.85, (255, 255, 255), 2, cv.LINE_AA)
cv.imshow("Text on image", image)
cv.waitKey(0)
cv.destroyAllWindows()
```

* **Draw Line and Rectangle on Image**

```python
image = cv.imread("Motor.jpg")
print(image.shape)
```

```python
image = cv.imread("Motor.jpg")

# Draw line
cv.line(image, (35, 267//2), (365, 267//2), (255, 255, 255), 2)

# Draw Rectangle
cv.rectangle(image, (25, 15), (365, 230), (255, 255, 255), 3)

cv.imshow("line & Rectangle", image)
cv.waitKey(0)
cv.destroyAllWindows()
```

* **Draw Circle and Ellipse on Image**

```python
image = cv.imread("W140.jpg")
print(image.shape)
```

```python
image = cv.imread("W140.jpg")

# Draw Circle
cv.circle(image, (310, 250), 150, (255, 255, 255), 2)

# Draw Ellipse
cv.ellipse(image, (150, 150), (50, 50), 100, 100, 200, (255, 255, 255), 2)

cv.imshow("Circle & Ellipse", image)
cv.waitKey(0)
cv.destroyAllWindows()
```

* **Draw Polygons on Image**

```python
arr = np.array([[50, 50], (20, 40), [50, 80], [80, 100]]).reshape((-1, 1, 2))

image = cv.imread('F22_Raptor.jpg')

cv.polylines(image, [arr], isClosed=True, color=(255, 255, 255), thickness=2)

cv.imshow("polygon", image)
cv.waitKey(0)
cv.destroyAllWindows()
```
