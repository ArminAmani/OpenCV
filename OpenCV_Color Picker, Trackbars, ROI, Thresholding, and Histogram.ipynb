* **Color Picker using Mouse Callback**

```python
def click_event(event, x, y, flags, params):
    if event == cv.EVENT_LBUTTONDOWN:
        b, g, r = img[y, x]
        print(f"Color at ({x}, {y} = R: {r}, G: {g}, B: {b})")

img = cv.imread("W140.jpg")
cv.imshow("Main image", img)
cv.setMouseCallback("Main image", click_event)
cv.waitKey(0)
cv.destroyAllWindows()
```

*â€Œ **Get trackbarpos**

```python
def nothing(x):
    pass

img = np.zeros((300, 512, 3), np.uint8)
cv.namedWindow("Trackbars")

cv.createTrackbar('R', 'Trackbars', 0, 255, nothing)
cv.createTrackbar('G', 'Trackbars', 0, 255, nothing)
cv.createTrackbar('B', 'Trackbars', 0, 255, nothing)

while True:
    r = cv.getTrackbarPos('R', 'Trackbars')
    g = cv.getTrackbarPos('G', 'Trackbars')
    b = cv.getTrackbarPos('B', 'Trackbars')

    img[:] = [b, g, r]
    cv.imshow('Trackbars', img)

    if cv.waitKey(1) & 0xFF == 27:
        break

cv.destroyAllWindows()
```

* **Flip, Rotate & Transpose**

```python
img_m = cv.imread("Motor.jpg")

flip_horizontal = cv.flip(img_m, 1)
flip_vertical = cv.flip(img_m, 0)
transpose = cv.transpose(img_m)

cv.imshow("main image", img_m)
cv.imshow("Horizontal image", flip_horizontal)
cv.imshow("Vertical image", flip_vertical)
cv.imshow("Transpose", transpose)

cv.waitKey(0)
cv.destroyAllWindows()
```

* **Saving a video**

```python
cap = cv.VideoCapture(0)
fourcc = cv.VideoWriter_fourcc(*'XVID')
out = cv.VideoWriter('output.avi', fourcc, 20.0, (640, 480))

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    out.write(frame)
    cv.imshow("Recording", frame)
    if cv.waitKey(1) & 0Xff == ord('q'):
        break

cap.release()
out.release()

cv.destroyAllWindows()
```

* **Color Spaces**

```python
F22_img = cv.imread("F22_Raptor.jpg")

gray = cv.cvtColor(F22_img, cv.COLOR_BGR2GRAY)
hsv = cv.cvtColor(F22_img, cv.COLOR_BGR2HSV)
lab = cv.cvtColor(F22_img, cv.COLOR_BGR2Lab)

cv.imshow("Original", F22_img)
cv.imshow("Gray", gray)
cv.imshow("HSV", hsv)
cv.imshow("LAB", lab)

cv.waitKey(0)
cv.destroyAllWindows()
```

* **Filter Color**

```python
cap = cv.VideoCapture(0)

while True:
    ret, frame = cap.read()
    hsv = cv.cvtColor(frame, cv.COLOR_BGR2HSV)

    # Define range of red color
    lower_red = np.array([0, 120, 70])
    upper_red = np.array([10, 255, 255])

    mask = cv.inRange(hsv, lower_red, upper_red)
    result = cv.bitwise_and(frame, frame, mask=mask)

    cv.imshow("Original", frame)
    cv.imshow("Mask", mask)
    cv.imshow("Filtered", result)

    if cv.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv.destroyAllWindows()
```

* **Perspective Transformation**

```python
image_c = cv.imread("W140.jpg")

pts1 = np.float32([[100, 100], [500, 100], [100, 400], [500, 400]])
pts2 = np.float32([[0, 0], [300, 0], [0, 300], [300, 300]])

matrics = cv.getPerspectiveTransform(pts1, pts2)
result = cv.warpPerspective(image_c, matrics, (100, 300))

cv.imshow("Original", image_c)
cv.imshow("Warped", result)

cv.waitKey(0)
cv.destroyAllWindows()
```

* **Simple Thresholding**

```python
image_m = cv.imread("Motor.jpg", 0)
_, thr = cv.threshold(image_m, 127, 255, cv.THRESH_BINARY)

cv.imshow("Original", image_m)
cv.imshow("Threshold", thr)
cv.waitKey(0)
cv.destroyAllWindows()
```

* **Otsu's Binarization Threshold**

```python
img_F22 = cv.imread("F22_Raptor.jpg", 0)
blur = cv.GaussianBlur(img_F22, (5, 5), 0)
_, otsu_thr = cv.threshold(blur, 0, 255, cv.THRESH_BINARY + cv.THRESH_OTSU)

cv.imshow("Original", img_F22)
cv.imshow("Blur", blur)
cv.imshow("Otsu Threshold", otsu_thr)

cv.waitKey(0)
cv.destroyAllWindows()
```
