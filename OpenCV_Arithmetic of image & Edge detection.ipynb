* **Addition and Subtraction**

```python
img1 = cv.imread("W140.jpg")
img2 = cv.imread("F22_Raptor.jpg")

img2 = cv.resize(img2, img1.shape[1::-1])

# Addition
added = cv.add(img1, img2)

# Subtraction
subtracted = cv.subtract(img1, img2)

cv.imshow("Addition", added)
cv.imshow("Subtract", subtracted)

cv.waitKey(0)
cv.destroyAllWindows()
```

* **Bitwise Operations on Binary Images**

```python
img1 = np.zeros((300, 300), dtype=np.uint8)
img2 = np.zeros((300, 300), dtype=np.uint8)

cv.rectangle(img1, (50, 50), (250, 250), 255, -1)
cv.circle(img2, (150, 150), 100, 255, -1)

bit_and = cv.bitwise_and(img1, img2)
bit_or = cv.bitwise_or(img1, img2)
bit_xor = cv.bitwise_xor(img1, img2)
bit_not2 = cv.bitwise_not(img2)

cv.imshow("and", bit_and)
cv.imshow("or", bit_or)
cv.imshow("xor", bit_xor)
cv.imshow("not img2", bit_not2)

cv.waitKey(0)
cv.destroyAllWindows()
```

* **Edge detection in OpenCV**

```python
img = cv.imread("F22_Raptor.jpg")

edge = cv.Canny(img, threshold1=40, threshold2=60)

cv.imshow('Edge', edge)
cv.waitKey(0)
cv.destroyAllWindows()
```

* **Scalling & Rotating**

```python
img = cv.imread('W140.jpg')

# scaling
scaled = cv.resize(img, (350, 350))

# Rotation
(h, w) = img.shape[:2]
matrics = cv.getRotationMatrix2D((w//2, h//2), 45, 1.0)
rotated = cv.warpAffine(img, matrics, (w, h))

cv.imshow("scaled image", scaled)
cv.imshow("Rotated image", rotated)
cv.waitKey(0)
cv.destroyAllWindows()
```

* **Image Blurring in OpenCV**

```python
image_M = cv.imread("Motor.jpg")

# GaussianBlur
gaussian = cv.GaussianBlur(image_M, (5, 5), 0)          # the kernel size for GaussianBlur must be positive and odd numbers.

# MedianBlur
median = cv.medianBlur(image_M, 3)

# Bilateral Filter
bilateral = cv.bilateralFilter(image_M, 8, 150, 150)


cv.imshow("Original Image", image_M)
cv.imshow("GaussianBlur", gaussian)
cv.imshow("MedianBlur", median)
cv.imshow("Bilateral Filter", bilateral)

cv.waitKey(0)
cv.destroyAllWindows()
```

* **Save Image with cv.imwrite()**

```python
# Grayscale
gray_image_c = cv.imread("W140.jpg", cv.IMREAD_GRAYSCALE)

cv.imwrite("GrayScaled_W140.jpg", gray_image_c)

print("Imaged saved successfully")
```

* **Add Border**

```python
F22_img = cv.imread("F22_Raptor.jpg")

bordered = cv.copyMakeBorder(F22_img, 5, 5, 5, 5, cv.BORDER_CONSTANT, value=(0, 0, 255))

cv.imshow("Bordered Image", bordered)
cv.waitKey(0)
cv.destroyAllWindows()
```

* **Play a Video**

```python
cap = cv.VideoCapture("FSI.mp4")

while cap.isOpened():
    ret, frame = cap.read()

    if not ret:
        break
    cv.imshow("video playback", frame)
    if cv.waitKey(5) & 0xFF == ord('q'):            # lower waitkey ====> faster video
        break

cap.release()
cv.destroyAllWindows()
```

* **Capture Video from the Camera**

```python
cap = cv.VideoCapture(0)

while True:
    ret, frame = cap.read()
    cv.imshow("Camera Feed", frame)
    if cv.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv.destroyAllWindows()
```
